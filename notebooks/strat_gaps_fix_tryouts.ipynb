# STRAT - Short Tandem Repeat Analysis Tool

## 1. Parameters

```python
path_ontarget = '/opt/data/workdir/bc3_1.guppy.ontarget.tsv'
path_processed = '/opt/data/workdir/bc3_1.guppy.ontarget.processed.tsv'

COLUMNS = [
    'direction',
    'id',
    'prefix_flank',
    'ins',
    'suffix_flank',
    'prefix_flank_q',
    'ins_q',
    'suffix_flank_q',
]
```

## 2. Imports

```python
from csv import QUOTE_NONE
import numpy as np
import pandas as pd
```

## 3. Tryouts

```python
dfo = pd.read_csv(path_ontarget, sep='\t', header=None, dtype=str, quoting=QUOTE_NONE)
dfo.columns = COLUMNS
```

```python
len(dfo)
```

```python
dfp = pd.read_csv(path_processed, sep='\t')
```

```python
len(dfp)
```

```python
cond = dfp['len_ins_ext_aln'] > 60
cond &= dfp['ins_ext_aln'].str.slice(0, 15) == 'CAGCAGCAGCAGCGG'
ids = set(dfp[cond]['id'])
len(ids)
```

```python
cond = dfo['id'].isin(ids)
cond &= dfo['direction'] == 'fwd'
dfo[cond].groupby(['prefix_flank']).count()
```

```python
cond = dfo['id'].isin(ids)
cond &= dfo['direction'] == 'rev'
print(len)
dfo[cond].groupby(['suffix_flank']).count()['id'].reset_index().sort_values('id', ascending=False)[:10]
```
