## Functions

```python
def count_kmers(seq):
    len_kmer = 11
    kmers = {}
    
    for i in range(len(seq) - len_kmer + 1):
        kmer = seq[i:i+len_kmer]
        if kmer in kmers:
            kmers[kmer].append(i)
        else:
            kmers[kmer] = [i]

    return {k: kmers[k][0] for k in kmers if len(kmers[k]) == 1}


def load_fastq_tsv(path):
    df = pd.read_csv(path, sep='\t', header=None)
    df.columns = ['seq', 'id']
    return df
```

```python
from collections import Counter

def get_top_5(kmers):
    merged_dict = Counter()

    for k in kmers:
        merged_dict.update(k)

    return sorted(merged_dict.items(), key=lambda x: x[1], reverse=True)[:5]


def count_kmers_fastq(seq):
    len_kmer = 11
    kmers = {}
    
    for i in range(len(seq) - len_kmer + 1):
        kmer = seq[i:i+len_kmer]
        if kmer in kmers:
            kmers[kmer].append(i)
        else:
            kmers[kmer] = [i]

    # return {k: kmers[k] for k in kmers if kmers[k] >= 1}
    return [
        {k: kmers[k][0] for k in kmers if (len(kmers[k]) == 1) and (k in fwd_pre_kmers)},
        {k: kmers[k][0] for k in kmers if (len(kmers[k]) == 1) and (k in fwd_suf_kmers)},
        {k: kmers[k][0] for k in kmers if (len(kmers[k]) == 1) and (k in rev_pre_kmers)},
        {k: kmers[k][0] for k in kmers if (len(kmers[k]) == 1) and (k in rev_suf_kmers)},
    ]
```

```python
res = []

for d in datasets:
    dataset = d['dataset']
    caller = d['caller']
    path = f'{fastq_tsv_path}{dataset}_{caller}.fastq.tsv'
    df_fastq = load_fastq_tsv(path)
    df_fastq['dataset'] = dataset
    df_fastq['caller'] = caller
    # reads = len(df_fastq)
    df_fastq['kmers'] = df_fastq.apply(lambda x: count_kmers_fastq(x['seq']), axis=1)
    df_fastq['fwd_pre_kmers_fq'] = df_fastq.apply(lambda x: x['kmers'][0], axis=1)
    df_fastq['fwd_suf_kmers_fq'] = df_fastq.apply(lambda x: x['kmers'][1], axis=1)
    df_fastq['rev_pre_kmers_fq'] = df_fastq.apply(lambda x: x['kmers'][2], axis=1)
    df_fastq['rev_suf_kmers_fq'] = df_fastq.apply(lambda x: x['kmers'][3], axis=1)

    res.append(df_fastq)
```

```python
import statistics

def crunch_kmers(row):
    fwd_cnt = 0
    fwd_mean = None
    # fwd_stdev = None
    if row['fwd_pre_kmers_fq'] and row['fwd_suf_kmers_fq']:
        fwd_deltas = []
        i = 0
        for kp, vp in row['fwd_pre_kmers_fq'].items():
            j = 0
            for ks, vs in row['fwd_suf_kmers_fq'].items():
                
                dist_flat = fwd_suf_kmers[ks] - fwd_pre_kmers[kp]
                dist_inst = vs - vp
                fwd_deltas.append(dist_inst - dist_flat)
                j += 1
                if j == 100:
                    break
            i += 1
            if i == 100:
                break
        fwd_cnt = len(fwd_deltas)
        try:
            fwd_mean = statistics.median(fwd_deltas)
            # fwd_stdev = statistics.stdev(fwd_deltas)
        except:
            print(row['fwd_pre_kmers_fq'])
            print(row['fwd_suf_kmers_fq'])
            raise

    rev_cnt = 0
    rev_mean = None
    # rev_stdev = None
    if row['rev_pre_kmers_fq'] and row['rev_suf_kmers_fq']:
        rev_deltas = []
        i = 0
        for kp, vp in row['rev_pre_kmers_fq'].items():
            j = 0
            for ks, vs in row['rev_suf_kmers_fq'].items():
                
                dist_flat = rev_suf_kmers[ks] - rev_pre_kmers[kp]
                dist_inst = vs - vp
                rev_deltas.append(dist_inst - dist_flat)
                j += 1
                if j == 100:
                    break
            i += 1
            if i == 100:
                break
        rev_cnt = len(rev_deltas)
        try:
            rev_mean = statistics.median(rev_deltas)
            # rev_stdev = statistics.stdev(rev_deltas)
        except:
            print(row['rev_pre_kmers_fq'])
            print(row['rev_suf_kmers_fq'])
            raise

    return pd.Series({
        'fwd_cnt': fwd_cnt,
        'fwd_mean': fwd_mean,
        # 'fwd_stdev': fwd_stdev,
        'rev_cnt': rev_cnt,
        'rev_mean': rev_mean,
        # 'rev_stdev': rev_stdev,
    })
```

```python
res = [pd.concat([r, r.apply(crunch_kmers, axis=1)], axis=1) for r in res]
```

```python
def crunch_starts_and_ends(row):
    fwd_starts = []
    fwd_ends = []
    rev_starts = []
    rev_ends = []
    if row['fwd_pre_kmers_fq']:
        i = 0
        for kp, vp in row['fwd_pre_kmers_fq'].items():
            fwd_starts.append(339 + vp - fwd_pre_kmers[kp])
    if row['fwd_suf_kmers_fq']:
        i = 0
        for ks, vs in row['fwd_suf_kmers_fq'].items():
            fwd_ends.append(339 + vs - fwd_suf_kmers[ks])
    if row['rev_pre_kmers_fq']:
        i = 0
        for kp, vp in row['rev_pre_kmers_fq'].items():
            rev_starts.append(335 + vp - rev_pre_kmers[kp])
    if row['rev_suf_kmers_fq']:
        i = 0
        for ks, vs in row['rev_suf_kmers_fq'].items():
            rev_ends.append(335 + vs - rev_suf_kmers[ks])

    fwd_start = None
    fwd_end = None
    rev_start = None
    rev_end = None
    
    fwd = False
    try:
        fwd_start = statistics.median(fwd_starts)
        fwd_end = statistics.median(fwd_ends)
        fwd = True
    except Exception:
        fwd = False

    rev = False
    try:
        rev_start = statistics.median(rev_starts)
        rev_end = statistics.median(rev_ends)
        rev = True
    except Exception:
        rev = False

    return pd.Series({
        'fwd': fwd,
        'fwd_start': fwd_start,
        'fwd_end': fwd_end,
        'rev': rev,
        'rev_start': rev_start,
        'rev_end': rev_end,
    })
```

```python
import numpy as np

def snap_insert(row):
    if row['fwd']:
        return row['seq'][int(np.floor(row['fwd_start'])):int(np.ceil(row['fwd_end']))]
    if row['rev']:
        return row['seq'][int(np.floor(row['rev_start'])):int(np.ceil(row['rev_end']))]

    return None
```
