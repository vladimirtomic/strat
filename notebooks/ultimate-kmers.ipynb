## Functions

```python
def count_kmers(seq):
    len_kmer = 11
    kmers = {}
    
    for i in range(len(seq) - len_kmer + 1):
        kmer = seq[i:i+len_kmer]
        if kmer in kmers:
            kmers[kmer].append(i)
        else:
            kmers[kmer] = [i]

    return {k: kmers[k][0] for k in kmers if len(kmers[k]) == 1}


def load_fastq_tsv(path):
    df = pd.read_csv(path, sep='\t', header=None)
    df.columns = ['seq', 'id']
    return df


def process_points(points, round):
    if not points:
        return [None, None, None]

    try:
        point = int(round(stats.median(points)))
        point_cnt = len(points)
        point_stdev = stats.stdev(points)
    except Exception as e:
        # raise e
        return [None, None, None]

    return [point, point_cnt, point_stdev]


def count_kmers_fastq(seq):
    len_kmer = 11
    kmers = {}
    
    for i in range(len(seq) - len_kmer + 1):
        kmer = seq[i:i+len_kmer]
        if kmer in kmers:
            kmers[kmer].append(i)
        else:
            kmers[kmer] = [i]

    fwd_starts = []
    fwd_ends = []
    rev_starts = []
    rev_ends = []
    
    for k in kmers:
        if len(kmers[k]) == 1:
            if k in FWD_PRE_KMERS:
                fwd_starts.append(339 + kmers[k][0] - FWD_PRE_KMERS[k])
            elif k in FWD_SUF_KMERS:
                fwd_ends.append(339 + kmers[k][0] - FWD_SUF_KMERS[k])
            elif k in REV_PRE_KMERS:
                rev_starts.append(335 + kmers[k][0] - REV_PRE_KMERS[k])
            elif k in REV_SUF_KMERS:
                rev_ends.append(335 + kmers[k][0] - REV_SUF_KMERS[k])
    
    fwd_start, fwd_start_cnt, fwd_start_stdev = process_points(fwd_starts, np.floor)
    fwd_end, fwd_end_cnt, fwd_end_stdev = process_points(fwd_ends, np.ceil)
    rev_start, rev_start_cnt, rev_start_stdev = process_points(rev_starts, np.floor)
    rev_end, rev_end_cnt, rev_end_stdev = process_points(rev_ends, np.ceil)

    return pd.Series({
        'fwd_start': fwd_start,
        'fwd_start_cnt': fwd_start_cnt,
        'fwd_start_stdev': fwd_start_stdev,
        'fwd_end': fwd_end,
        'fwd_end_cnt': fwd_end_cnt,
        'fwd_end_stdev': fwd_end_stdev,
        'rev_start': rev_start,
        'rev_start_cnt': rev_start_cnt,
        'rev_start_stdev': rev_start_stdev,
        'rev_end': rev_end,
        'rev_end_cnt': rev_end_cnt,
        'rev_end_stdev': rev_end_stdev,
    })


def snap_insert_fwd(row):
    if row['fwd'] == 1:
        return pd.Series({
            'prefix_flank': row['seq'][int(row['fwd_start'])-10:int(row['fwd_start'])],
            'ins': row['seq'][int(row['fwd_start']):int(row['fwd_end'])],
            'suffix_flank': row['seq'][int(row['fwd_end']):int(row['fwd_end'])+10],
        })

    return None, None, None


def snap_insert_rev(row):
    if row['rev'] == 1:
        return pd.Series({
            'prefix_flank': row['seq'][int(row['rev_start'])-10:int(row['rev_start'])],
            'ins': row['seq'][int(row['rev_start']):int(row['rev_end'])],
            'suffix_flank': row['seq'][int(row['rev_end']):int(row['rev_end'])+10],
        })

    return None, None, None
```

## Main

```python
FWD_PRE_KMERS = count_kmers(fwd_flank_pre)
FWD_SUF_KMERS = count_kmers(fwd_flank_suf)
REV_PRE_KMERS = count_kmers(rev_flank_pre)
REV_SUF_KMERS = count_kmers(rev_flank_suf)

fwd_pre_kmers_set = set(count_kmers(fwd_flank_pre).keys())
fwd_suf_kmers_set = set(count_kmers(fwd_flank_suf).keys())
rev_pre_kmers_set = set(count_kmers(rev_flank_pre).keys())
rev_suf_kmers_set = set(count_kmers(rev_flank_suf).keys())
```

```python
res = []

for d in datasets:
    dataset = d['dataset']
    caller = d['caller']
    path = f'{fastq_tsv_path}{dataset}_{caller}.fastq.tsv'
    df_fastq = load_fastq_tsv(path)
    df_fastq['dataset'] = dataset
    df_fastq['caller'] = caller
    # reads = len(df_fastq)
    df_fastq = pd.concat([df_fastq, df_fastq.apply(lambda x: count_kmers_fastq(x['seq']), axis=1)], axis=1)
    
    df_fastq['fwd'] = ((df_fastq['fwd_start'] > 0) & (df_fastq['fwd_end'] > df_fastq['fwd_start'])).astype(int)
    df_fastq['rev'] = ((df_fastq['rev_start'] > 0) & (df_fastq['rev_end'] > df_fastq['rev_start'])).astype(int)

    cond = df_fastq['fwd'] == 1
    dff = df_fastq[cond].copy()
    dff['direction'] = 'fwd'
    dff = pd.concat([dff, dff.apply(snap_insert_fwd, axis=1)], axis=1)
    dff = dff[['direction', 'id', 'prefix_flank', 'ins', 'suffix_flank', 'fwd_start_cnt', 'fwd_start_stdev', 'fwd_end_cnt', 'fwd_end_stdev']]
    dff.columns = ['direction', 'id', 'prefix_flank', 'ins', 'suffix_flank', 'start_cnt', 'start_stdev', 'end_cnt', 'end_stdev']
    
    cond = df_fastq['rev'] == 1
    dfr = df_fastq[cond].copy()
    dfr['direction'] = 'rev'
    dfr = pd.concat([dfr, dfr.apply(snap_insert_rev, axis=1)], axis=1)
    dfr = dfr[['direction', 'id', 'prefix_flank', 'ins', 'suffix_flank', 'rev_start_cnt', 'rev_start_stdev', 'rev_end_cnt', 'rev_end_stdev']]
    dfr.columns = ['direction', 'id', 'prefix_flank', 'ins', 'suffix_flank', 'start_cnt', 'start_stdev', 'end_cnt', 'end_stdev']

    df = pd.concat([dff, dfr])
    output_path = f'{kmers_tsv_path}{dataset}.{caller}.kmers.tsv'
    df.sort_values(['id', 'direction']).to_csv(output_path, sep='\t', index=False, header=False)
    
    res.append(df_fastq)
```
